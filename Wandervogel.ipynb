```python
# Import wichtiger Python-Module für Datenanalyse

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame, Series

import re
```

```python
# Einlesen der Wandervogel-Metadaten
wandervogel_df = pd.read_csv('metadatenwandervogelf5000.csv')

# Einlesen der Liste der Koordinaten von Schweizer Ortschaften
koord_df = pd.read_csv('PLZO_CSV_LV03.csv', sep=';')
```

```python
# Sample Liste Wandervogel
wandervogel_df
```

```python
# Sample Liste Ortschaften
koord_df
```

```python
# Das genauste Datum scheint immer das Datum nach dem letzten Strichpunkt zu sein. Die These soll kurz überprüft werden.
# Doch als erstes mässen alle NaNs im DataFrame durch leere Strings ersetzt werden.
wandervogel_df.fillna('', inplace=True)

set([l[-1] for l in wandervogel_df.Zeit.str.split(';')])
```

```python
# Die These scheint zu stimmen. Uns interessiert nur die genauste Zeitangabe,
# also ersetzen den jeweiligen String durch diese
wandervogel_df.Zeit.replace('.*;', '', regex=True, inplace=True)
```

```python
# Als nächstes soll die Ortsspalte in der Wandervogel-Tabelle in Kontinent, Land, Kanton und Ort aufgeteilt werden.
# Als Trenner zwischen diesen Entitäten dient offenbar der Strichpunkt.
# Nicht alle Einträge haben offenbar alle vier Kategorien Kontinent, Land, Kanton, Ort. Dies muss verbessert werden,
# zuerst soll aber noch überprüft werden, ob es Einträge gibt, welche mehr als drei Trenner-; haben
vector = [len(l) > 4 for l in wandervogel_df.Ort.str.split(';')]
wandervogel_df[vector]
```

```python
# Diese Einträge müssen bereinigt werden, indem der letzte Teil gelöscht wird
wandervogel_df.replace(' ?/Europa;S.*$', '', regex=True, inplace=True)
```

```python
# Wir erstellen ein neues Datenframe mit den Orten nach Kontinent, Land, Kanton und Ort getrennt
orte = pd.DataFrame(data=wandervogel_df.Ort.str.split(';').tolist(), columns=['Kontinent', 'Land', 'Kanton', 'Ort'])
orte.fillna('', inplace=True)
```

```python
# Als nächstes überprüfen wir, ob es noch weitere Unregelmässigkeiten in den Ortsdaten gibt. Dazu erstellen wir eine
# Liste der vorhandenen Ortsangaben
print(orte.Kontinent.unique())
print(orte.Land.unique())
print(orte.Kanton.unique())
print(orte.Ort.unique())
```

```python
# Die Länder werden durch ihre Kürzel ersetzt
laender_map = {'Schweiz': 'CH', 'Deutschland': 'GER', 'Italien': 'IT', 'Frankreich': 'FRA'}
for k, v in laender_map.items():
    orte.Land.replace(k, v, inplace=True)
```

```python
# Bei den Kantonen scheint es zwei Probleme zu geben: Einerseits das Postfix ", Kanton", anderseits den Schrägstrich
# gefolgt allenfalls von "Europa".
orte.Kanton.replace(', Kanton$', '', inplace=True, regex=True)
orte.Kanton.replace(' /.*$', '', inplace=True, regex=True)
```

```python
# Die Kantone werden ebenfalls durch ihre Kürzel ersetzt
kantone_map = {'Zürich': 'ZH', 'Thurgau': 'TG', 'Bern': 'BE', 'Zug': 'ZG', 'Aargau': 'AG',
               'Glarus': 'GL', 'Appenzell Ausserrhoden': 'AR', 'Baselland': 'BL', 'Jura': 'JU',
              'Baselstadt': 'BS', 'Neuchâtel':'NE', 'St. Gallen': 'SG', 'Luzern': 'LU', 'Solothurn': 'SO',
              'Schaffhausen': 'SH', 'Obwalden': 'OW', 'Graubünden': 'GR', 'Tessin': 'TI',
               'Wallis': 'VS', 'Nidwalden': 'NW', 'Uri': 'UR', 'Schwyz': 'SZ'}
for k, v in kantone_map.items():
    orte.Kanton.replace(k, v, inplace=True)
```

```python
# Bei den Ortschaften scheint es folgende Probleme zu geben: Schrägstriche, ", Stadt"-Zusatz, Kantons-Zusatz (in Klammern)
# sehr kurze Namen sowie abgeschnittene Namen
orte.Ort.replace(' /.*$', '', inplace=True, regex=True)
orte.Ort.replace(', Stad.*$', '', inplace=True, regex=True)
orte.Ort.replace(' \(.*$', '', inplace=True, regex=True)
```

```python
# Bevor das Problem der sehr kurzen und abgeschnittenen Namen angegangen wird, sollen die beiden Datenframes wieder zu einem
# vereinigt werden
wandervogel_df.drop('Ort', axis=1, inplace=True)
wandervogel_neu = pd.concat([wandervogel_df, orte], axis=1)
wandervogel_neu
```

```python
# Das Problem der sehr kurzen Namen
vector = [(len(l) < 3) and (len(l) > 0) for l in wandervogel_neu['Ort']]
wandervogel_neu[vector]
```

```python
# Orte werden manuell verbessert
wandervogel_neu.iat[130,6] = 'Schaffhausen'
wandervogel_neu.iat[131,6] = 'Schaffhausen'
wandervogel_neu.iat[158,6] = 'Mühlhausen'
wandervogel_neu.iat[498,6] = 'Rüdlingen'
```

```python
# Zeige Ortschaften an, die einen partiellen Match im Titel haben (Hinweis auf eine abgetrennte Bezeichnung)
def_column_width= pd.options.display.max_colwidth
def_rows_displayed = pd.options.display.max_rows
pd.options.display.max_colwidth = 140
pd.options.display.max_rows = 100
vector = [True if re.search(o+'\w+', t) and o is not '' else False for t, o in zip(wandervogel_neu.Titel, wandervogel_neu.Ort)]
wandervogel_neu.loc[vector, ['Titel', 'Ort']]
pd.options.display.max_colwidth = def_column_width
pd.options.display.max_rows = def_rows_displayed
```

```python
orte_ignoriert = [462, 463, 464]


for i, e in enumerate(vector):
    if e and i not in orte_ignoriert:
        wandervogel_neu['Ort'].iat[i] = re.search(wandervogel_neu['Ort'].iat[i]+'\w+', wandervogel_neu['Titel'].iat[i]).group(0)
```

```python
orte_ungueltig = [150, 151, 152, 153, 154, 156, 161, 198, 229, 275, 276, 325, 326, 373 ]

for i in orte_ungueltig:
    wandervogel_neu['Kanton'].iat[i] = ''
    wandervogel_neu['Ort'].iat[i] = ''
```

```python
orte_ersetzt = {145: 'Bisital', 146: 'Fluhberg', 147: 'Schwarzenberg', 148: 'Schwarzenberg',
               155: 'Thurmündung', 162: 'Pfannenstiel', 163: 'Seelis', 226: 'Engadin', 227: 'Engadin',
               228: 'Engadin', 230: 'Bözberg', 231: 'Morcote', 232: 'Hinter-Goldingen', 237: 'Lämmerengletscher',
               334: 'Sonnenberg', 569: 'Speer', 570: 'Speer', 571: 'Speer', 572: 'Speer'}

for k, v in orte_ersetzt.items():
    wandervogel_neu['Kanton'].iat[k] = ''
    wandervogel_neu['Ort'].iat[k] = v
```

```python
koords = DataFrame(columns=['E', 'N'])
empty_df = DataFrame(columns=['E', 'N'])
empty_df.loc[0] = ['', '']

for i, (kanton, ort) in enumerate(zip(wandervogel_neu.Kanton, wandervogel_neu.Ort)):
    vector = [(ort == o) and (kanton == k) for o, k in zip(koord_df.Ortschaftsname, koord_df.Kantonskürzel)]
    koord_temp = koord_df.loc[vector, ['E', 'N']]
    koords = pd.concat([koords, koord_temp[0:1] if koord_temp.shape[0] > 0 else empty_df], axis=0)
wandervogel_final = pd.concat([wandervogel_neu, koords.reset_index(drop=True)], axis=1)
```

```python
wandervogel_final.to_csv('output.csv', encoding='utf-8')
```
